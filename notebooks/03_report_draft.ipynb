## Introduction

Biodegradability is an important property in environmental chemistry because it reflects whether a chemical compound can be broken down efficiently in the environment. Compounds with poor biodegradability may persist for longer periods, increasing the risk of environmental accumulation and potential harm.

Experimental evaluation of biodegradability can be time-consuming and resource-intensive. For this reason, computational approaches are widely used to support early-stage screening. Among these approaches, Quantitative Structure–Activity Relationship (QSAR) methods aim to predict chemical properties or behaviors from numerical descriptors derived from molecular structure.

In this project, the QSAR biodegradation dataset is used to study whether molecular descriptors can be used to predict whether a compound is readily biodegradable. The analysis combines exploratory data analysis and baseline machine learning models. First, several interpretable descriptors are examined to understand class-related distribution patterns. Then, three baseline classifiers — Logistic Regression, Decision Tree, and Random Forest — are trained and compared. Finally, model interpretation is used to connect the predictive results with descriptor importance.

The results show that biodegradability prediction depends on a combination of multiple molecular descriptors rather than a single feature alone. In particular, both interpretable descriptors and more abstract QSAR descriptors contribute to classification performance.

## Dataset and Methods

This project uses the QSAR biodegradation dataset, which includes 1055 compounds and 41 molecular descriptors. The classification target indicates whether a compound is readily biodegradable (RB) or not readily biodegradable (NRB). For clarity, the original labels were remapped from numeric codes to descriptive class names during preprocessing.

The dataset was loaded in Python using `fetch_openml`. The anonymous feature names were replaced with the original descriptor names, and the feature matrix was combined with the mapped target labels to form a single analysis table. Basic inspection was then carried out to examine the dataset structure and check for missing values.

The study was conducted in three stages. First, exploratory data analysis was performed to examine class distribution, descriptive statistics, boxplots of selected interpretable descriptors, and pairwise correlations. A preliminary screening procedure was also used to identify candidate important descriptors across all 41 variables.

Second, baseline classification models were developed. The data were split into training and test sets using an 80/20 stratified split with a fixed random seed. Three baseline models were trained and evaluated: Logistic Regression, Decision Tree, and Random Forest.

Third, model interpretation was performed through confusion-matrix-based error analysis and Random Forest feature importance. The model-based important features were then compared with the EDA-based screening results to assess consistency between exploratory findings and predictive modeling.

## Results

### Class Distribution

The QSAR biodegradation dataset is moderately imbalanced, containing 699 NRB samples and 356 RB samples. Although the imbalance is not extreme, it suggests that model evaluation should consider not only overall accuracy but also class-specific performance, especially for the RB class.

### Exploratory Data Analysis Results

The exploratory analysis showed that several interpretable descriptors display meaningful differences between the two classes. In particular, nO tends to be higher in RB, whereas nN, nX, and C% tend to be higher in NRB. These results suggest that individual descriptors contain useful class-related information, although no single descriptor appears sufficient for perfect separation.

The numerical comparison of selected descriptors further confirmed these patterns. For example, RB showed higher mean and median values of nO, while NRB showed higher values of nN, nX, and C%. Correlation analysis among selected interpretable descriptors indicated that most feature pairs were only weakly correlated, suggesting that these variables may provide complementary rather than highly redundant information.

A preliminary screening across all 41 descriptors identified several candidate important features, including SpMax_L, SpMax_A, SpPosA_B(p), SM6_B(m), HyWi_B(m), SM6_L, and SpMax_B(m). This suggests that biodegradability classification is influenced not only by easily interpretable descriptors, but also by more abstract QSAR descriptors.

### Baseline Model Performance

Three baseline classifiers were trained and evaluated: Logistic Regression, Decision Tree, and Random Forest. Their performance is summarized below.

- Logistic Regression achieved an accuracy of 0.858, with F1-scores of 0.892 for class 0 and 0.792 for class 1.
- Decision Tree achieved an accuracy of 0.815, with F1-scores of 0.863 for class 0 and 0.715 for class 1.
- Random Forest achieved the highest accuracy of 0.877, with F1-scores of 0.912 for class 0 and 0.797 for class 1.

Overall, Random Forest performed best in terms of overall classification accuracy, while Logistic Regression remained highly competitive.

### Error Analysis

Confusion-matrix-based error analysis revealed important differences among the three models. Random Forest correctly classified the largest number of NRB samples and produced the fewest NRB-to-RB errors, which explains its strong overall accuracy. However, Logistic Regression correctly identified more RB samples and produced the fewest RB-to-NRB errors, giving it the highest recall for the RB class.

In contrast, Decision Tree showed the weakest performance overall and missed the largest number of true RB samples. This suggests that a single default decision tree is less stable than the other two baseline models.

### Feature Importance and Consistency with EDA

Random Forest feature importance showed that the final prediction relies most strongly on descriptors such as SpMax_B(m), SM6_B(m), SpMax_L, SpPosA_B(p), and SpMax_A. Many of these features were also identified in the earlier EDA-based screening, indicating good consistency between preliminary statistical analysis and model-based interpretation.

Specifically, seven descriptors — HyWi_B(m), SM6_B(m), SM6_L, SpMax_A, SpMax_B(m), SpMax_L, and SpPosA_B(p) — appeared in both the EDA top 10 and the Random Forest top 10. At the same time, a few descriptors appeared only in the EDA ranking or only in the Random Forest ranking, suggesting that some features are informative individually, while others become more important when nonlinear interactions are considered by the model.

### Overall Result

Overall, the results suggest that biodegradability prediction depends on a combination of multiple molecular descriptors rather than any single feature alone. Random Forest is the strongest overall baseline model in this project, while Logistic Regression remains particularly valuable when sensitivity to the RB class is important.

## Discussion

The results of this project suggest that molecular descriptors can provide useful information for predicting biodegradability. Even a simple baseline model such as Logistic Regression achieved relatively strong performance, which indicates that the QSAR descriptors in this dataset already contain meaningful class-related patterns. At the same time, the stronger overall performance of Random Forest suggests that some nonlinear relationships and feature interactions may also play an important role.

The exploratory analysis and the model-based interpretation were broadly consistent. Several descriptors identified in the preliminary EDA screening also appeared among the most important features in the Random Forest model. This strengthens the reliability of the analysis, because the important patterns observed in simple statistical comparisons were not completely lost when the data were modeled using a more complex classifier. However, the model also highlighted additional descriptors that were not among the strongest features in the earlier mean-difference screening, which suggests that some useful predictive information only becomes visible in a multivariate and nonlinear setting.

Another important finding is that model quality depends on the evaluation metric being considered. Random Forest achieved the highest overall accuracy, but Logistic Regression achieved the highest recall for the RB class. This means that the choice of “best” model depends partly on the practical goal. If overall classification performance is the main priority, Random Forest is the strongest baseline in this project. If identifying as many RB compounds as possible is more important, Logistic Regression remains highly competitive.

This project also has several limitations. First, only baseline models were used, and no hyperparameter tuning was performed. Second, the current analysis relied on precomputed descriptors, which means the project did not explore alternative molecular representations. Third, the dataset is moderately imbalanced, which may influence model behavior, especially for the RB class. In addition, feature importance from Random Forest shows relative contribution but does not directly explain causal chemical mechanisms.

Overall, the project demonstrates that biodegradability prediction is feasible with standard QSAR descriptors and baseline machine learning methods. However, it also suggests that stronger predictive performance and deeper scientific interpretation would likely require further feature engineering, model tuning, and more advanced representation methods.

## Conclusion and Future Work

This project investigated whether molecular descriptors from the QSAR biodegradation dataset can be used to predict whether a compound is readily biodegradable. The analysis combined exploratory data analysis, baseline classification models, and model interpretation.

The results show that the dataset contains meaningful class-related information. Several interpretable descriptors, such as nO, nN, nX, and C%, showed useful differences between RB and NRB in the EDA stage. At the same time, model-based analysis showed that several abstract QSAR descriptors, including SpMax_B(m), SM6_B(m), SpMax_L, and SpPosA_B(p), played important roles in the final classification.

Among the three baseline models, Random Forest achieved the best overall performance, while Logistic Regression remained competitive and achieved the highest recall for the RB class. These results suggest that biodegradability prediction depends on a combination of multiple descriptors rather than any single feature alone.

There are several directions for future work. First, model performance could be improved through hyperparameter tuning and additional validation strategies. Second, feature selection methods could be applied to identify a smaller but more effective subset of descriptors. Third, alternative molecular representations and more advanced models could be explored to improve predictive performance and scientific interpretability. Finally, further work could focus on developing a clearer chemical interpretation of the most important descriptors identified by the model.

Overall, this project provides a useful baseline study of QSAR-based biodegradability prediction and establishes a foundation for further improvement in both modeling and interpretation.
